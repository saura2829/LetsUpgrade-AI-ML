# Basic plot 


```python
import matplotlib.pyplot as plt

import matplotlib
```

```python
#Declaring the array

x = np.array(credit_card['Time'])

y = np.array(credit_card['Amount'])

#Ploting the Graph
print(plt.plot(x,y))


#Labling the Graph
print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

# Box Plot

```python
credit_card.boxplot(column='Amount',figsize=(10,10))
```

# Bar Plot

```python
# Declearing the arrays 
a = np.array(credit_card.head(100)['Time'])
b = np.array(credit_card.head(100)['Amount'])

# Ploting the graph
print(plt.bar(a,b))

# Labling the graph
print(plt.xlabel('Amount'))
print(plt.ylabel('Time'))
print(plt.title('Amount VS Time'))
```

# 7. Histogram to plot different columns
